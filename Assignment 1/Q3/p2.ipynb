```python
from solcx import compile_standard
from web3 import Web3
import json
```

```python
contract_path = "/Users/subhu/Desktop/Sem/Sem_7/BlockChain/BlockChain/Assignment 1/Q3/20CS10064.sol"
with open(contract_path, 'r') as file:
    contract_source_code = file.read()

compiled_sol = compile_standard({
    "language": "Solidity",
    "sources": { "20CS10064.sol": {
        "content": contract_source_code
        }},
    "settings": {
        "outputSelection": {
            "*": { "*": ["abi", "metadata", "evm.bytecode", "evm.sourceMap" ] }
        }
    }
}, solc_version= "0.8.0")
```

```python
with open("compiled_TicketBooking.json", "w") as file:
    json.dump(compiled_sol, file)
```

```python
infura_url = "https://sepolia.infura.io/v3/d32544fba98b4074bb60a9635b4b1121"
w3 = Web3(Web3.HTTPProvider(infura_url))
```

```python
chain_id = 11155111
wallet_addr = "0x9b3710f9b284eAC355F1721B00C381c626250439"
nonce = w3.eth.get_transaction_count(wallet_addr)
acc_private_key = "0x22751dbfeddda2597c6ae0fbb3008ec8dd9f5f3902d29410fa84c3d96f73ffb1"
```

```python
abi = compiled_sol['contracts']['20CS10064.sol']['TicketBooking']['abi']
bytecode = compiled_sol['contracts']['20CS10064.sol']['TicketBooking']['evm']['bytecode']['object']

TicketBooking = w3.eth.contract(abi=abi, bytecode=bytecode)
```

```python
TicketBooking_txn = TicketBooking.constructor(1000, 500).build_transaction(
    {
        "gas": 2000000,
        "gasPrice": w3.eth.gas_price,
        "chainId": chain_id,
        "nonce": nonce
    }
)
```

```python
signed_txn = w3.eth.account.sign_transaction(TicketBooking_txn, private_key=acc_private_key )
print(signed_txn)
```

```python
txn_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)
print(txn_hash)
```

```python
txn_receipt = w3.eth.wait_for_transaction_receipt(txn_hash)
print(txn_receipt)
```
